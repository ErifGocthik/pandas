```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(df['pH'], df['Grade'], c='green', s=10, marker='^')
plt.xlabel('pH')
plt.ylabel('Качество')
plt.show()
```

```python
low_grade = df[df['Grade'] == 0]  # получаем строки, в которых Grade равно нулю
medium_grade = df[df['Grade'] == 1]
high_grade = df[df['Grade'] == 2]
low_grade
```

```python
plt.scatter(low_grade['pH'], low_grade['Grade'], c='red', s=10, marker='^')
plt.scatter(medium_grade['pH'], medium_grade['Grade'], c='blue', s=20, marker='+')
plt.scatter(high_grade['pH'], high_grade['Grade'], c='green', s=10)
plt.xlabel('pH')
plt.ylabel('Качество')
plt.show()
```

```python
import matplotlib
```

```python
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(projection='3d')

ax.scatter(df['pH'], df['Colour'], df['Grade'], c='green', marker='^')
ax.set_xlabel('pH')
ax.set_ylabel('Color')
ax.set_zlabel('Grade')
plt.show()
```

```python
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(projection='3d')

ax.scatter(low_grade['pH'], low_grade['Colour'], low_grade['Grade'], c='red', marker='^')
ax.scatter(medium_grade['pH'], medium_grade['Colour'], medium_grade['Grade'], c='blue', marker='1')
ax.scatter(high_grade['pH'], high_grade['Colour'], high_grade['Grade'], c='green')
ax.set_xlabel('pH')
ax.set_ylabel('Color')
ax.set_zlabel('Grade')
plt.show()
```
