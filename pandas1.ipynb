```python
plt.scatter(df['pH'], df['Fat'], c='green', s=50, marker='1')
plt.scatter(df['Odor'], df['Taste'], c='blue', s=50, marker='1')
plt.xlabel('pH')
plt.ylabel('Качество')
plt.show()
```

```python
low_grade = df[df['Grade'] == 0]
medium_grade = df[df['Grade'] == 1]
high_grade = df[df['Grade'] == 2]
high_grade
```

```python
plt.scatter(low_grade['pH'], low_grade['Grade'], c = 'red', s = 50, marker = '3')
plt.scatter(medium_grade['pH'], medium_grade['Grade'], c = 'blue', s = 60, marker = '1')
plt.scatter(high_grade['pH'], high_grade['Grade'], c = 'green', s = 60, marker = '2')
plt.xlabel('pH')
plt.ylabel('Качество')
plt.show()
```
